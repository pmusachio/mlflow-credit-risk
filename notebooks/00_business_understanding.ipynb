# MLflow Credit Risk - Business Understanding

This notebook documents the business problem, assumptions, success metric, and analytical plan.

## Business Problem

A credit department needs to classify default risk with traceable experiments.

## Objective

Classify default risk with a reproducible pipeline, versioned metrics, and optional MLflow tracking.

## Primary metric

ROC AUC, average precision, F1, and optional MLflow run logging

## Business Assumptions

- Credit decisions require reproducible metrics and a model history.
- False negatives can increase defaults; false positives can reduce approvals for good customers.
- MLflow helps compare model and parameter versions.

## Solution Strategy

1. **Step 01. Data Description:** Validate schema, dimensions, missing values, types, and granularity.
2. **Step 02. Feature Engineering:** Create domain-oriented variables for the challenge.
3. **Step 03. Data Filtering:** Remove records with no analytical value or leakage risk.
4. **Step 04. Exploratory Data Analysis:** Validate hypotheses and separate relevant signal from noise.
5. **Step 05. Data Preparation:** Impute, scale, and encode variables for modeling.
6. **Step 06. Feature Selection:** Separate IDs, target, input variables, and dropped columns.
7. **Step 07. Machine Learning Modelling:** Train a reproducible baseline and evaluate technical metrics.
8. **Step 08. Hyperparameter/Fit Strategy:** Reserve room for tuning, threshold choice, and model comparison.
9. **Step 09. Business Translation:** Convert metrics into decisions, prioritization, risk, revenue, or operations.
10. **Step 10. Delivery:** Version experiments in MLflow, promote the model to Production, and document the API/SageMaker.
