# MLflow Credit Risk - Exploratory Analysis

Notebook para validar hipoteses e gerar graficos para storytelling.

```python
from pathlib import Path
import sys
import json

PROJECT_ROOT = Path.cwd().resolve()
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from mlflow_credit_risk.config import load_config
from mlflow_credit_risk.data import load_training_frame, profile_dataframe
from mlflow_credit_risk.features import prepare_features

config = load_config()
config["project"]
```

```python
df = load_training_frame(config)
df.describe(include='all').T.head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

target = config.get('data', {}).get('target')
if target:
    target = target.lower()
    if target in df.columns:
        ax = df[target].value_counts(normalize=True).head(20).plot(kind='bar', figsize=(10, 4), title=f'Distribuicao de {target}')
        ax.set_ylabel('share')
        plt.show()
```

```python
numeric = df.select_dtypes(include='number')
if not numeric.empty:
    cols = numeric.columns[:8]
    numeric[cols].hist(figsize=(14, 8), bins=30)
    plt.tight_layout()
    plt.show()
```
